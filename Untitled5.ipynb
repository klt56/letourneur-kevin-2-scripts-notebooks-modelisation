```python
import re
import tensorflow as tf

# ------- Chargement du modèle SavedModel -------
MODEL_DIR = "./models/bilstm_stemming_v1"
model = tf.keras.models.load_model(MODEL_DIR)
print("✅ Model loaded from", MODEL_DIR)

# ------- Pré-traitement (stemming + stopwords) -------
# Dépendances :
#   pip install nltk
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

# Stopwords NLTK (téléchargement si nécessaire)
try:
    _ = stopwords.words("english")
except LookupError:
    nltk.download("stopwords", quiet=True)

stop_words = set(stopwords.words("english"))

# On garde les négations (important en sentiment)
NEGATIONS = {"no", "nor", "not", "never"}
stop_words = stop_words - NEGATIONS

# Stopwords spécifiques tweets (comme dans l’étape 2)
stop_words |= {"rt", "amp", "im", "dont", "u", "ur", "ive", "youre", "thats"}

stemmer = PorterStemmer()
token_re = re.compile(r"[a-z]+")

def preprocess_stem(text: str, min_len: int = 2) -> str:
    """
    Nettoyage tweet + stopwords (en gardant les négations) + stemming.
    Le modèle a été entraîné sur ce format, donc il faut appliquer la même logique en production.
    """
    if not isinstance(text, str):
        return ""

    text = text.lower()
    text = re.sub(r"http\S+|www\.\S+", " ", text)   # URLs
    text = re.sub(r"@\w+", " ", text)              # mentions
    text = re.sub(r"#", " ", text)                 # hashtags
    text = re.sub(r"[^a-z\s]", " ", text)          # lettres uniquement

    tokens = token_re.findall(text)
    tokens = [t for t in tokens if len(t) >= min_len and t not in stop_words]
    tokens = [stemmer.stem(t) for t in tokens]

    return " ".join(tokens)

def predict_sentiment(text: str):
    """
    Retour :
      label : 0 négatif / 1 positif
      proba_pos : probabilité de positif (sigmoid)
    """
    cleaned = preprocess_stem(text)

    # Le modèle Keras attend une entrée string (batch, 1)
    x = tf.constant([[cleaned]], dtype=tf.string)

    # Sortie sigmoid : proba de la classe positive
    proba_pos = float(model(x, training=False).numpy().reshape(-1)[0])
    label = 1 if proba_pos >= 0.5 else 0
    return label, proba_pos
```

```python
lbl, p = predict_sentiment("this flight was horrible i hate it")
print(lbl, p)  # lbl=0 attendu (négatif)
```
