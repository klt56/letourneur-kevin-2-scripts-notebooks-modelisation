```python
#NLTK : stopwords + lemmatisation + stemming
import re
import nltk
from nltk.corpus import stopwords, wordnet
from nltk.stem import WordNetLemmatizer, PorterStemmer
from functools import lru_cache

for pkg in ["stopwords", "wordnet", "omw-1.4"]:
    try:
        nltk.download(pkg, quiet=True)
    except Exception:
        pass

TAGGER_OK = False
for pkg in ["averaged_perceptron_tagger_eng", "averaged_perceptron_tagger"]:
    try:
        nltk.download(pkg, quiet=True)
        TAGGER_OK = True
        break
    except Exception:
        pass

stop_words = set(stopwords.words("english"))
NEGATIONS = {"no", "nor", "not", "never"}
stop_words = stop_words - NEGATIONS
stop_words |= {"rt", "amp", "im", "dont", "u", "ur", "ive", "youre", "thats"}

lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()
token_re = re.compile(r"[a-z]+")

def _map_pos(tag: str):
    if tag.startswith("J"): return wordnet.ADJ
    if tag.startswith("V"): return wordnet.VERB
    if tag.startswith("N"): return wordnet.NOUN
    if tag.startswith("R"): return wordnet.ADV
    return wordnet.NOUN

def pos_tag_safe(tokens):
    if not TAGGER_OK:
        return [(t, "NN") for t in tokens]
    try:
        from nltk import pos_tag
        try:
            return pos_tag(tokens, lang="eng")
        except TypeError:
            return pos_tag(tokens)
    except Exception:
        return [(t, "NN") for t in tokens]

@lru_cache(maxsize=200_000)
def _lemma_cached(word: str, wn_pos):
    return lemmatizer.lemmatize(word, wn_pos)

def clean_base(text: str, min_len: int = 2):
    if not isinstance(text, str):
        return []
    text = text.lower()
    text = re.sub(r"http\S+|www\.\S+", " ", text)
    text = re.sub(r"@\w+", " ", text)
    text = re.sub(r"#", " ", text)
    text = re.sub(r"[^a-z\s]", " ", text)
    tokens = token_re.findall(text)
    tokens = [t for t in tokens if len(t) >= min_len and t not in stop_words]
    return tokens

def preprocess_lemmatize(text: str) -> str:
    tokens = clean_base(text)
    if tokens:
        tagged = pos_tag_safe(tokens)
        tokens = [_lemma_cached(w, _map_pos(pos)) for w, pos in tagged]
    return " ".join(tokens)

def preprocess_stem(text: str) -> str:
    tokens = clean_base(text)
    if tokens:
        tokens = [stemmer.stem(w) for w in tokens]
    return " ".join(tokens)

print("TAGGER_OK =", TAGGER_OK)
print("✅ NLP preprocessing OK")
```

```python
#ensembles de données TF + vectoriseur + graphiques + suivi MLflow
from sklearn.metrics import accuracy_score, roc_auc_score, roc_curve
import mlflow
from src.mlflow_utils import setup_mlflow, track_run, Timer

setup_mlflow("air-paradis")

def make_tf_dataset(texts, labels, batch_size=64, shuffle=False):
    ds = tf.data.Dataset.from_tensor_slices((texts, labels))
    if shuffle:
        ds = ds.shuffle(buffer_size=min(len(texts), 20000), seed=RANDOM_STATE, reshuffle_each_iteration=True)
    ds = ds.batch(batch_size).prefetch(tf.data.AUTOTUNE)
    return ds

def make_vectorizer(train_texts, vocab_size=20000, seq_len=128):
    # TextVectorization sort une séquence d’IDs entiers ; le nettoyage est fait avant (lemma/stem)
    vec = tf.keras.layers.TextVectorization(
        max_tokens=vocab_size,
        output_mode="int",
        output_sequence_length=seq_len
    )
    vec.adapt(train_texts)
    vocab = vec.get_vocabulary()
    return vec, vocab

def plot_and_save_history(history, out_path):
    # History Keras : loss/auc/accuracy sur train et val
    os.makedirs(os.path.dirname(out_path), exist_ok=True)
    fig = plt.figure(figsize=(8, 4))
    for k, v in history.history.items():
        plt.plot(v, label=k)
    plt.title("Training history")
    plt.xlabel("Epoch")
    plt.ylabel("Value")
    plt.legend()
    plt.tight_layout()
    fig.savefig(out_path, dpi=160, bbox_inches="tight")
    plt.close(fig)

def plot_and_save_roc(y_true, y_score, out_path, title):
    os.makedirs(os.path.dirname(out_path), exist_ok=True)
    fpr, tpr, _ = roc_curve(y_true, y_score)
    fig = plt.figure(figsize=(6, 4))
    plt.plot(fpr, tpr)
    plt.plot([0, 1], [0, 1], linestyle="--")
    plt.title(title)
    plt.xlabel("FPR")
    plt.ylabel("TPR")
    plt.tight_layout()
    fig.savefig(out_path, dpi=160, bbox_inches="tight")
    plt.close(fig)

def evaluate_keras(model, X_eval, y_eval, batch_size=256):
    # Sortie probas (sigmoid)
    with Timer() as tp:
        y_score = model.predict(X_eval, batch_size=batch_size, verbose=0).reshape(-1)
    y_pred = (y_score >= 0.5).astype(int)
    acc = accuracy_score(y_eval, y_pred)
    auc = roc_auc_score(y_eval, y_score)
    return acc, auc, float(tp.dt), y_score

def train_log_keras(run_name, model, train_ds, X_val_text, y_val, params, artifacts_dir="mlartifacts_step2"):
    os.makedirs(artifacts_dir, exist_ok=True)

    callbacks = [
        tf.keras.callbacks.EarlyStopping(monitor="val_auc", mode="max", patience=2, restore_best_weights=True),
        tf.keras.callbacks.ReduceLROnPlateau(monitor="val_auc", mode="max", factor=0.5, patience=1, verbose=1),
    ]

    with Timer() as tfitt:
        history = model.fit(train_ds, validation_data=make_tf_dataset(X_val_text, y_val, batch_size=128), epochs=8, callbacks=callbacks, verbose=1)

    # Éval Validation
    val_acc, val_auc, pred_time, val_scores = evaluate_keras(model, X_val_text, y_val)

    # Artifacts
    hist_path = os.path.join(artifacts_dir, f"{run_name}_history.png")
    roc_path  = os.path.join(artifacts_dir, f"{run_name}_roc_val.png")
    plot_and_save_history(history, hist_path)
    plot_and_save_roc(y_val, val_scores, roc_path, title=f"ROC - {run_name} (Validation)")

    metrics = {
        "val_accuracy": float(val_acc),
        "val_auc": float(val_auc),
        "fit_time_sec": float(tfitt.dt),
        "predict_time_val_sec": float(pred_time)
    }

    run_id = track_run(
        run_name=run_name,
        params=params,
        metrics=metrics,
        artifacts={"history": hist_path, "roc_val": roc_path}
    )

    # Stockage du modèle dans MLflow
    with mlflow.start_run(run_id=run_id):
        try:
            mlflow.keras.log_model(model, artifact_path="model")
        except Exception:
            # fallback
            model.save(os.path.join(artifacts_dir, f"{run_name}_saved_model"), include_optimizer=False)
            mlflow.log_artifacts(os.path.join(artifacts_dir, f"{run_name}_saved_model"), artifact_path="saved_model")

    print(f"✅ {run_name} | val_acc={val_acc:.4f} | val_auc={val_auc:.4f} | run_id={run_id}")
    return {"run_name": run_name, "run_id": run_id, "val_acc": val_acc, "val_auc": val_auc}
```

```python
#Build BiLSTM (fonction)
def build_bilstm_model(vectorizer, vocab_size, embed_dim=100, lstm_units=64, dropout=0.3):
    inputs = tf.keras.Input(shape=(1,), dtype=tf.string, name="text")
    x = vectorizer(inputs)
    x = tf.keras.layers.Embedding(input_dim=vocab_size, output_dim=embed_dim, name="emb")(x)
    x = tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(lstm_units, return_sequences=False), name="bilstm")(x)
    x = tf.keras.layers.Dropout(dropout)(x)
    outputs = tf.keras.layers.Dense(1, activation="sigmoid", name="out")(x)

    model = tf.keras.Model(inputs, outputs)
    model.compile(
        optimizer=tf.keras.optimizers.Adam(1e-3),
        loss="binary_crossentropy",
        metrics=[tf.keras.metrics.BinaryAccuracy(name="accuracy"), tf.keras.metrics.AUC(name="auc")]
    )
    return model
```

```python
#Installer / importer Gensim (Word2Vec)
import sys
import subprocess

try:
    import gensim
    from gensim.models import Word2Vec
except Exception:
    subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", "gensim"])
    import gensim
    from gensim.models import Word2Vec

print("Gensim version:", gensim.__version__)
print("✅ Gensim OK")
```

```python
#Préparer tokens + entraîner Word2Vec sur Train
def to_token_lists(texts):
    # Les textes sont déjà nettoyés ; split simple
    return [t.split() for t in texts]

train_tokens = to_token_lists(X_train_best)

w2v_dim = 100
w2v = Word2Vec(
    sentences=train_tokens,
    vector_size=w2v_dim,
    window=5,
    min_count=2,
    workers=4,
    sg=1,              # skip-gram
    seed=RANDOM_STATE,
    epochs=10
)

print("✅ Word2Vec entraîné | vocab:", len(w2v.wv))
```

```python
#Construire embedding matrix Word2Vec alignée au vocab du TextVectorization
def build_embedding_matrix_from_gensim(vocab_list, gensim_wv, embed_dim):
    # vocab_list vient de TextVectorization.get_vocabulary()
    matrix = np.random.normal(scale=0.06, size=(len(vocab_list), embed_dim)).astype(np.float32)
    hits = 0
    for i, word in enumerate(vocab_list):
        if word in gensim_wv:
            matrix[i] = gensim_wv[word]
            hits += 1
    print(f"Embedding hits: {hits}/{len(vocab_list)}")
    return matrix

vec_best, vocab_best = make_vectorizer(X_train_best, vocab_size=VOCAB_SIZE, seq_len=SEQ_LEN)
emb_matrix_w2v = build_embedding_matrix_from_gensim(vocab_best, w2v.wv, w2v_dim)
```

```python
#Modèle BiLSTM avec Word2Vec (embedding figé)
def build_bilstm_with_matrix(vectorizer, emb_matrix, lstm_units=64, dropout=0.3, trainable=False):
    inputs = tf.keras.Input(shape=(1,), dtype=tf.string, name="text")
    x = vectorizer(inputs)
    emb = tf.keras.layers.Embedding(
        input_dim=emb_matrix.shape[0],
        output_dim=emb_matrix.shape[1],
        weights=[emb_matrix],
        trainable=trainable,
        name="emb_pretrained"
    )
    x = emb(x)
    x = tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(lstm_units), name="bilstm")(x)
    x = tf.keras.layers.Dropout(dropout)(x)
    outputs = tf.keras.layers.Dense(1, activation="sigmoid", name="out")(x)

    model = tf.keras.Model(inputs, outputs)
    model.compile(
        optimizer=tf.keras.optimizers.Adam(1e-3),
        loss="binary_crossentropy",
        metrics=[tf.keras.metrics.BinaryAccuracy(name="accuracy"), tf.keras.metrics.AUC(name="auc")]
    )
    return model

model_w2v = build_bilstm_with_matrix(vec_best, emb_matrix_w2v, lstm_units=64, dropout=0.3, trainable=False)
train_ds_best = make_tf_dataset(X_train_best, y_train, batch_size=BATCH_SIZE, shuffle=True)

res_w2v = train_log_keras(
    run_name="step2_bilstm_embedding_word2vec",
    model=model_w2v,
    train_ds=train_ds_best,
    X_val_text=X_val_best,
    y_val=y_val,
    params={
        "model": "BiLSTM",
        "preprocess": best_preproc + "+stopwords",
        "embedding": "Word2Vec_trained",
        "embedding_dim": w2v_dim,
        "embedding_trainable": False,
        "vocab_size": VOCAB_SIZE,
        "seq_len": SEQ_LEN,
        "lstm_units": 64,
        "dropout": 0.3,
        "batch_size": BATCH_SIZE,
        "sample_size": 16000
    }
)
```

```python
#Télécharger GloVe (automatique) + charger vecteurs
import zipfile
import urllib.request

GLOVE_DIR = "embeddings"
os.makedirs(GLOVE_DIR, exist_ok=True)

# GloVe 6B 100d
glove_zip = os.path.join(GLOVE_DIR, "glove.6B.zip")
glove_txt = os.path.join(GLOVE_DIR, "glove.6B.100d.txt")

if not os.path.exists(glove_txt):
    if not os.path.exists(glove_zip):
        url = "https://nlp.stanford.edu/data/glove.6B.zip"
        print("Téléchargement GloVe:", url)
        urllib.request.urlretrieve(url, glove_zip)
        print("✅ ZIP téléchargé:", glove_zip)

    print("Extraction ZIP...")
    with zipfile.ZipFile(glove_zip, "r") as z:
        z.extractall(GLOVE_DIR)
    print("✅ Extraction OK")

print("GloVe file:", glove_txt)

def load_glove(glove_path, embed_dim=100):
    glove = {}
    with open(glove_path, "r", encoding="utf-8") as f:
        for line in f:
            parts = line.rstrip().split(" ")
            word = parts[0]
            vec = np.asarray(parts[1:], dtype=np.float32)
            if vec.shape[0] == embed_dim:
                glove[word] = vec
    return glove

glove_dim = 100
glove = load_glove(glove_txt, embed_dim=glove_dim)
print("✅ GloVe loaded | vocab:", len(glove))
```

```python
#Embedding matrix GloVe + BiLSTM (embedding figé)
def build_embedding_matrix_from_glove(vocab_list, glove_dict, embed_dim):
    matrix = np.random.normal(scale=0.06, size=(len(vocab_list), embed_dim)).astype(np.float32)
    hits = 0
    for i, word in enumerate(vocab_list):
        if word in glove_dict:
            matrix[i] = glove_dict[word]
            hits += 1
    print(f"Embedding hits: {hits}/{len(vocab_list)}")
    return matrix

emb_matrix_glove = build_embedding_matrix_from_glove(vocab_best, glove, glove_dim)

model_glove = build_bilstm_with_matrix(vec_best, emb_matrix_glove, lstm_units=64, dropout=0.3, trainable=False)

res_glove = train_log_keras(
    run_name="step2_bilstm_embedding_glove",
    model=model_glove,
    train_ds=train_ds_best,
    X_val_text=X_val_best,
    y_val=y_val,
    params={
        "model": "BiLSTM",
        "preprocess": best_preproc + "+stopwords",
        "embedding": "GloVe_6B_100d",
        "embedding_dim": glove_dim,
        "embedding_trainable": False,
        "vocab_size": VOCAB_SIZE,
        "seq_len": SEQ_LEN,
        "lstm_units": 64,
        "dropout": 0.3,
        "batch_size": BATCH_SIZE,
        "sample_size": 16000
    }
)
```

```python
#Choix du meilleur embedding (sur AUC validation)
best_embedding = "glove" if res_glove["val_auc"] >= res_w2v["val_auc"] else "word2vec"
print("✅ Meilleur embedding (selon AUC val):", best_embedding)
```

```python
#Build + train CNN 1D (embedding random)
def build_cnn1d_model(vectorizer, vocab_size, embed_dim=100, filters=128, kernel_size=5, dropout=0.3):
    inputs = tf.keras.Input(shape=(1,), dtype=tf.string, name="text")
    x = vectorizer(inputs)
    x = tf.keras.layers.Embedding(input_dim=vocab_size, output_dim=embed_dim, name="emb")(x)
    x = tf.keras.layers.Conv1D(filters=filters, kernel_size=kernel_size, activation="relu")(x)
    x = tf.keras.layers.GlobalMaxPooling1D()(x)
    x = tf.keras.layers.Dropout(dropout)(x)
    outputs = tf.keras.layers.Dense(1, activation="sigmoid")(x)

    model = tf.keras.Model(inputs, outputs)
    model.compile(
        optimizer=tf.keras.optimizers.Adam(1e-3),
        loss="binary_crossentropy",
        metrics=[tf.keras.metrics.BinaryAccuracy(name="accuracy"), tf.keras.metrics.AUC(name="auc")]
    )
    return model

vec_cnn, vocab_cnn = make_vectorizer(X_train_best, vocab_size=VOCAB_SIZE, seq_len=SEQ_LEN)
model_cnn = build_cnn1d_model(vec_cnn, vocab_size=len(vocab_cnn), embed_dim=100, filters=128, kernel_size=5, dropout=0.3)

res_cnn = train_log_keras(
    run_name="step2_cnn1d_random_embedding",
    model=model_cnn,
    train_ds=train_ds_best,
    X_val_text=X_val_best,
    y_val=y_val,
    params={
        "model": "CNN1D",
        "preprocess": best_preproc + "+stopwords",
        "embedding": "random_trainable",
        "embedding_dim": 100,
        "vocab_size": VOCAB_SIZE,
        "seq_len": SEQ_LEN,
        "filters": 128,
        "kernel_size": 5,
        "dropout": 0.3,
        "batch_size": BATCH_SIZE,
        "sample_size": 16000
    }
)
```

```python
#Tokenization + datasets TF pour BERT
MODEL_NAME = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

MAX_LEN = 128
BERT_BATCH = 16

def tokenize_texts(texts):
    return tokenizer(
        list(texts),
        padding=True,
        truncation=True,
        max_length=MAX_LEN,
        return_tensors="tf"
    )

# Pour BERT, on utilise le texte brut (sans stemming/lemma), c’est le standard
enc_train = tokenize_texts(X_train)
enc_val   = tokenize_texts(X_val)
enc_test  = tokenize_texts(X_test)

train_ds_bert = tf.data.Dataset.from_tensor_slices((dict(enc_train), y_train)).shuffle(20000, seed=RANDOM_STATE).batch(BERT_BATCH)
val_ds_bert   = tf.data.Dataset.from_tensor_slices((dict(enc_val), y_val)).batch(BERT_BATCH)
test_ds_bert  = tf.data.Dataset.from_tensor_slices((dict(enc_test), y_test)).batch(BERT_BATCH)

print("✅ BERT datasets OK")
```

```python
#Fine-tuning BERT + MLflow (history + ROC)
bert = TFAutoModelForSequenceClassification.from_pretrained(MODEL_NAME, num_labels=2)

bert.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=2e-5),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=[tf.keras.metrics.SparseCategoricalAccuracy(name="accuracy")]
)

with Timer() as tfit:
    hist_bert = bert.fit(train_ds_bert, validation_data=val_ds_bert, epochs=1, verbose=1)

# Probas validation pour AUC (logits -> softmax -> proba classe 1)
with Timer() as tp:
    logits_val = bert.predict(val_ds_bert, verbose=0).logits
    proba_val_pos = tf.nn.softmax(logits_val, axis=-1)[:, 1].numpy()

y_val_pred = (proba_val_pos >= 0.5).astype(int)

from sklearn.metrics import accuracy_score, roc_auc_score

val_acc = accuracy_score(y_val, y_val_pred)
val_auc = roc_auc_score(y_val, proba_val_pos)

os.makedirs("mlartifacts_step2", exist_ok=True)
hist_path = "mlartifacts_step2/bert_history.png"
roc_path  = "mlartifacts_step2/bert_roc_val.png"

plot_and_save_history(hist_bert, hist_path)
plot_and_save_roc(y_val, proba_val_pos, roc_path, title="ROC - BERT (Validation)")

run_id = track_run(
    run_name="step2_bert_finetune",
    params={
        "model": MODEL_NAME,
        "max_len": MAX_LEN,
        "batch_size": BERT_BATCH,
        "epochs": 1,
        "sample_size": 16000
    },
    metrics={
        "val_accuracy": float(val_acc),
        "val_auc": float(val_auc),
        "fit_time_sec": float(tfit.dt),
        "predict_time_val_sec": float(tp.dt)
    },
    artifacts={"history": hist_path, "roc_val": roc_path}
)

with mlflow.start_run(run_id=run_id):
    try:
        mlflow.keras.log_model(bert, artifact_path="model")
    except Exception:
        bert.save("mlartifacts_step2/bert_saved_model", include_optimizer=False)
        mlflow.log_artifacts("mlartifacts_step2/bert_saved_model", artifact_path="saved_model")

print("✅ BERT | val_acc:", val_acc, "| val_auc:", val_auc, "| run_id:", run_id)
```
